```python
#Carga desde un archivo .csv sin indice
df= pd.read_csv("Ventas_totales.csv")
df
```

```python
df.info()
```

```python
#Ver los valores nulos por columnas
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
df1=df.copy()
df1
```

Considerando que estamos hablando de las ventas es que preferi usar una medida de sustituci√≥n aritm√©tica


```python
df1['salon_ventas']=df1['salon_ventas'].fillna(round(df['salon_ventas'].median(),1))
valores_nulos1=df1.isnull().sum()
valores_nulos1
```

Considerando que este es un subtotal de las ventas es que prefer√≠ usar la misma de la pasada considerando que son fluctuantes las ventas entre los dias


```python
df1['subtotal_ventas_alimentos_bebidas']=df1['subtotal_ventas_alimentos_bebidas'].fillna(round(df['subtotal_ventas_alimentos_bebidas'].mean(),1))
valores_nulos2=df1.isnull().sum()
valores_nulos2
```

Considerando que son m√°s los valores faltantes es que usare la media de la muestra apra llenarlos y no ver tan afectado el modelo


```python
df1['tarjetas_credito']=df1['tarjetas_credito'].fillna(round(df['tarjetas_credito'].mean(),1))
valores_nulos3=df1.isnull().sum()
valores_nulos3
```

Considerando que estas columnas son valores menores a 3 valores faltantes es que creo que debamos de usar una medida de media o mediana considerando que es solo un valor que puede que este afecte el comportamiento real de la variable por lo cual es que usare el usar el valor anterior o posterior al mismo.
